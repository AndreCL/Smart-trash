```python
# In detail graphs data
def get_av(list_time, list2, time_frame):
    temp_time = list_time[-1]
    counter = 0
    total = 0
    timestamp_aver = []
    timestamp_aver.append(temp_time)
    temp1 = []
    temp2 = []
    for i in reversed(list_time):
        difference = i - temp_time
        if difference.seconds < time_frame: #3600 = 1h, 7200 = 2h, etc.
            total = total + list2[list_time.index(i)]
            counter = counter + 1
        else:
            temp1.append(total/counter)
            temp2.append(counter)
            counter = 1
            total = list2[list_time.index(i)]
            temp_time = i
            timestamp_aver.append(temp_time)

    if counter > 0:
        temp1.append(total/counter)
        temp2.append(counter)
    
    return temp1, temp2, timestamp_aver
```
